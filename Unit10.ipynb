```r
diab_scaled <- diabetes
```

```r
smp_size <- floor(0.75 * nrow(diab_scaled))
```

```r
set.seed(1337)
train_ind <- sample(seq_len(nrow(diab_scaled)), size = smp_size)
```

```r
train <- diab_scaled[train_ind, ]
test <- diab_scaled[-train_ind, ]
train_data <- dplyr::select(train, -readmitted)
train_labels <- train$readmitted
test_data <- dplyr::select(test, -readmitted)
test_labels <- test$readmitted
```

```r
model <- keras_model_sequential() %>%
  layer_dense(units = 10, activation = "relu", input_shape = c(69)) %>%
  layer_dense(units = 2, activation = "softmax")
```

```r
# Print a summary of a model
summary(model)
```

```r
# Compile the model
model %>% compile(optimizer = "rmsprop",loss = "binary_crossentropy", metrics = c("accuracy"))
```

```r
history <- model %>% fit(x_train, y_train, epochs = 5, batch_size = 512)
```
